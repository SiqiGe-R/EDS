# Understanding and Explanation

Table 3 estimates a linear probability model where the outcome is an indicator for being married (vs. cohabiting). The key regressor is the marriage subsidy in thousands of dollars. OLS is problematic because couples may adjust labour supply in response to marriage or taxes (simultaneity bias) and because reported income is measured with error. The IV strategy addresses both issues: predicted earnings, based only on pre-treatment 2012 characteristics, are used to construct a predicted subsidy that is orthogonal to post-treatment labour supply decisions. The control function specification (columns 5–6) adds a fifth-order polynomial in predicted earnings and predicted earnings split to ensure identification is driven by tax law variation rather than any spurious correlation between earnings levels and marriage propensity. The estimated coefficients imply that financial incentives have a statistically significant but economically modest impact on marriage decisions.

# Shortcomings

The main limitation is the first stage LASSO. Our first-stage F-statistics range from 15–32, compared to 15–60 in the paper. This stems from two sources: our LASSO feature set omits college major codes (a strong earnings predictor), and our threshold calibration achieves employment shares of 0.957/0.923 versus the paper's 0.963/0.969, compressing predicted earnings and weakening the instrument. As a result, our IV estimates may be less precisely estimated than those in the original paper, making direct comparison more challenging.

A second limitation arises from the approximation of tax inputs. Several income components — including self-employment income, capital gains, and retirement distributions — are either omitted or imperfectly measured in the ACS relative to administrative tax records. Because marriage penalties are larger for couples with non-wage income, our subsidy estimates are likely biased toward zero for high-income couples, partially explaining level differences in Table 2 relative to the original paper.

A further potential limitation relates to the validity of the instrumental variable strategy. The approach assumes that predicted income is exogenous to marriage decisions. However, predicted income is constructed using observable characteristics such as education, occupation, and demographics. If these characteristics are also correlated with underlying preferences for marriage, the exclusion restriction could potentially be violated. While the paper mitigates this concern by including rich controls and using a simulated IV approach, some residual correlation may still remain.

Our sample contains 37,907 couples versus the paper's 37,234, likely due to minor differences in relationship-status cleaning rules, though this does not materially affect results.



# Opportunities for Extensions

**Heterogeneous effects by earnings quantile using quantile IV regression.** The paper reports that marriage elasticities decline with household income (Figure 4), but uses a parametric polynomial interaction. An extension would estimate quantile IV regressions to obtain distributional treatment effects non-parametrically, testing whether the marriage response is concentrated among households at specific points of the earnings distribution rather than smoothly declining.

**Impact of the 2018 Tax Cuts and Jobs Act on different-sex couples.** The paper simulates TCJA effects for same-sex couples (Figure 5). The same machinery — LASSO income prediction + TAXSIM + marriage subsidy calculation — could be applied to different-sex cohabiting couples using ACS 2018–2022 data. This would test whether the TCJA-induced increase in marriage subsidies for high earners actually increased marriage rates among cohabiting heterosexual couples, providing a direct test of the paper's out-of-sample prediction with a much larger sample.

**Alternative machine learning methods for income prediction.** The paper uses LASSO to predict earnings. An extension would compare alternative approaches such as random forests or gradient boosting to assess whether improved prediction accuracy strengthens the first-stage relationship and affects IV estimates. This would help evaluate how sensitive the estimated marriage elasticity is to the choice of prediction model, and whether the results are driven by modelling assumptions rather than underlying economic relationships.
